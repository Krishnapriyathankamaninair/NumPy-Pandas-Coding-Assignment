```python
import numpy as np
import pandas as pd

salaries = np.array([45000, 52000, 61000, 48000, 75000])
# Increase all salaries by 10%.
increased_salary=(salaries*.1)+salaries
print("Salaries after 10% increase:",
increased_salary)
```

```python
# Find the average salary
average_salary = np.mean(salaries)
print("Average Salary :",average_salary )
avr_inc_salary=np.mean(increased_salary)
print("Average Salary after increment:",avr_inc_salary )
```

```python
#Find maximum and minimum salary
max_salary = np.max(salaries)
min_salary = np.min(salaries)

print("Maximum Salary:",max_salary)

print("Minimum Salary:",min_salary)
```

```python
#rint salaries greater than 60000
greater_salaries = salaries[salaries > 60000]
print("Salaries greater than 60000:",greater_salaries)

```

Q2. Sales Data Analysis

```python
sales = np.array([120, 150, 180, 200, 170, 160, 210, 220, 190, 175])
#Compute total sales
total_sales = np.sum(sales)
print("Total Sales:",total_sales)


#Compute mean and standard deviation
mean_sales = np.mean(sales)
std_sales = np.std(sales)

print("Mean Sales:",mean_sales)
print("Standard Deviation of Sales:",std_sales)


# Identify days where sales exceeded the average

days_above_avg = np.where(sales > mean_sales)[0] + 1

print("Days where sales exceeded average :",days_above_avg)



#Normalize using Min-Max scaling
min_sales = np.min(sales)
max_sales = np.max(sales)

normalized_sales = (sales - min_sales) / (max_sales - min_sales)

print('Normalized Sales:', normalized_sales)


```

Q3. Matrix Operations for Business Data

```python
sales_matrix = np.array([
[120,140,130,150],
[80,90,100,95],
[200,210,190,205]
])


# Total sales per product
product_sales = np.sum(sales_matrix, axis=1)
print("Total Sales per Product:",product_sales)


#Total sales per region
region_sales = np.sum(sales_matrix, axis=0)
print("Total Sales per Region:",region_sales)


#Identify the best performing product
best_product_index = np.argmax(product_sales)
print("Best Performing Product :", 'product',best_product_index+1)



# Identify the region with highest sales
best_region_index = np.argmax(region_sales)
print("Region with Highest Sales :", 'Region',best_region_index+1)

```

Q4:Employee Dataset Exploration

```python
data = {
'Employee':['A','B','C','D','E','F'],
'Department':['HR','IT','IT','Finance','HR','Finance'],
'Salary':[50000,65000,70000,72000,52000,68000],
'Experience':[2,5,7,6,3,4]
}

#Convert dictionary into DataFrame
df = pd.DataFrame(data)

print(df)



# Display basic statistics
print("\nBasic Statistics:")
print(df.describe())



#Find average salary per department
avg_salary_dept = df.groupby('Department')['Salary'].mean()
print("\nAverage Salary per Department:",'\n',avg_salary_dept)




# Identify employees with experience > 4 years
experienced_employees = df[df['Experience'] > 4]

print("\nEmployees with Experience > 4 years:",'\n', experienced_employees)


#Add Bonus column (10% of salary)
df['Bonus'] = df['Salary'] * 0.10

print("\nDataFrame with Bonus Column:")
print(df)

```

Q5. Missing Data Handling

```python
data = {
'Name':['John','Anna','Peter','Linda','David'],
'Age':[28, np.nan, 35, 32, np.nan],
'Salary':[50000,54000,np.nan,58000,60000]
}
df = pd.DataFrame(data)

# Detect missing values
print("\nMissing Values:")
print(df.isnull())

print("\nCount of Missing Values per Column:")
print(df.isnull().sum())




#Replace missing Age with mean age
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)

print("\nAfter Filling Missing Age with Mean:")
print(df)



# Replace missing Salary with median salary
median_salary = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(median_salary)

print("\nAfter Filling Missing Salary with Median:")
print(df)


#  Remove rows with any remaining missing values
df_cleaned = df.dropna()

print("\nFinal Cleaned DataFrame:")
print(df_cleaned)
```

Q6. Sales Performance Analysis

```python
sales_data = {
'Region':['North','South','East','West','North','South','East','West'],
'Salesperson':['A','B','C','D','E','F','G','H'],
'Sales':[12000,15000,13000,17000,16000,14000,15500,16500]
}

df = pd.DataFrame(sales_data)


# Total sales per region
region_sales = df.groupby('Region')['Sales'].sum()

print("\nTotal Sales per Region:")
print(region_sales)


#Identify the top salesperson
top_salesperson = df.loc[df['Sales'].idxmax()]

print("\nTop Salesperson:")
print(top_salesperson)



#Compute average sales across regions
avg_sales_region = df.groupby('Region')['Sales'].mean()

print("\nAverage Sales per Region:")
print(avg_sales_region)

#  Sort dataset by sales descending
sorted_df = df.sort_values(by='Sales', ascending=False)

print("\nDataset Sorted by Sales Descending:")
print(sorted_df)
```

Q7. Customer Purchase Analysis

```python
data = {
'CustomerID':[101,102,103,104,105,106,107],
'Age':[25,45,35,50,23,40,60],
'PurchaseAmount':[250,600,400,800,150,700,900],
'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}

df = pd.DataFrame(data)


#Average purchase amount by city
avg_purchase_city = df.groupby('City')['PurchaseAmount'].mean()
print("\nAverage Purchase Amount by City:")
print(avg_purchase_city)



#Customers with purchase above overall average
overall_avg = df['PurchaseAmount'].mean()
above_avg_customers = df[df['PurchaseAmount'] > overall_avg]
print("\nCustomers with Purchase Above Average:")
print(above_avg_customers)



# City generating highest revenue
city_revenue = df.groupby('City')['PurchaseAmount'].sum()
top_city = city_revenue.idxmax()


print("\nCity with Highest Revenue:")
print(top_city)


# Create Age_Group column
df['Age_Group'] = np.where(df['Age'] < 30, 'Young',
                   np.where(df['Age'] <= 50, 'Adult', 'Senior'))

print("\nDataFrame with Age Group:")
print(df)



# Average purchase per Age Group
avg_purchase_group = df.groupby('Age_Group')['PurchaseAmount'].mean()

print("\nAverage Purchase per Age Group:")
print(avg_purchase_group)
```

Q8. Time Series Sales Analysis

```python
dates = pd.date_range('2024-01-01', periods=12, freq='M')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date':dates,'Sales':sales})
print(df)

# Convert Date to datetime index
df.set_index('Date', inplace=True)

print("\nDataFrame with Date as Index:")
print(df)



# Compute monthly percentage growth
df['Monthly_Growth (%)'] = df['Sales'].pct_change() * 100

print("\nMonthly Percentage Growth:")
print(df)

# Calculate 3-month rolling average sales
df['Rolling_Avg_3M'] = df['Sales'].rolling(window=3).mean()

print("\n 3-Month Rolling Average:")
print(df)


# Identify month with highest growth rate
max_growth_month = df['Monthly_Growth (%)'].idxmax()
max_growth_value = df['Monthly_Growth (%)'].max()

print("\nMonth with Highest Growth Rate:")
print(max_growth_month)

print("\nHighest Growth Percentage:")
print(max_growth_value)

```
